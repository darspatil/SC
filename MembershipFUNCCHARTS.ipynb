## Membership Functions(types and implementations)
in fuzzy logic a membership function (MF) defines how much a given input belongs to a fuzzy set typically on scale of 0 to 1.

There are various types of membership functions

1)Triangular MF

2)Trapezoidal MF

3)Gaussian MF

4)Generalized Bell MF

5)Sigmoidal MF

6)S function MF

7)Constant MF



# 1) Traingular MF

```python
import numpy as np
import matplotlib.pyplot as plt

def triangular_mf(x, a, b, c):
    return np.maximum(np.minimum((x - a) / (b - a), (c - x) / (c - b)), 0)

# Define x-axis range
x = np.linspace(0, 10, 1000)

# Define the triangle parameters
a, b, c = 2, 5, 8

# Compute membership values
y = triangular_mf(x, a, b, c)

# Plot
plt.plot(x, y, label=f'Triangular MF ({a}, {b}, {c})', color='blue')
plt.title('Triangular Membership Function')
plt.xlabel('Input')
plt.ylabel('Membership Degree')
plt.grid(True)
plt.legend()
plt.show()
```

# 2)Trapazoidal MF

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the trapezoidal membership function
def trapezoidal_mf(x, a, b, c, d):
    return np.maximum(np.minimum(np.minimum((x - a)/(b - a), 1), (d - x)/(d - c)), 0)

# Define the x-axis range
x = np.linspace(0, 12, 1000)

# Trapezoid parameters
a, b, c, d = 2, 4, 8, 10

# Compute membership values
y = trapezoidal_mf(x, a, b, c, d)

# Plot the function
plt.plot(x, y, label=f'Trapezoidal MF ({a}, {b}, {c}, {d})', color='green')
plt.title('Trapezoidal Membership Function')
plt.xlabel('Input')
plt.ylabel('Membership Degree')
plt.grid(True)
plt.legend()
plt.show()
```

# 3)Gaussian MF

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Gaussian membership function
def gaussian_mf(x, c, sigma):
    return np.exp(-((x - c) ** 2) / (2 * sigma ** 2))

# Define x-axis range
x = np.linspace(0, 12, 1000)

# Gaussian parameters
c = 6       # Center of the bell curve
sigma = 1.5 # Standard deviation (width)

# Compute membership values
y = gaussian_mf(x, c, sigma)

# Plot
plt.plot(x, y, label=f'Gaussian MF (c={c}, ={sigma})', color='purple')
plt.title('Gaussian Membership Function')
plt.xlabel('Input')
plt.ylabel('Membership Degree')
plt.grid(True)
plt.legend()
plt.show()
```

# 4)Generalized Bell MF

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Generalized Bell Membership Function
def gbell_mf(x, a, b, c):
    return 1 / (1 + np.abs((x - c) / a) ** (2 * b))

# Define the x-axis range
x = np.linspace(0, 12, 1000)

# Parameters for the bell curve
a = 2     # Width
b = 4     # Slope
c = 6     # Center

# Compute membership values
y = gbell_mf(x, a, b, c)

# Plot
plt.plot(x, y, label=f'Generalized Bell MF (a={a}, b={b}, c={c})', color='orange')
plt.title('Generalized Bell Membership Function')
plt.xlabel('Input')
plt.ylabel('Membership Degree')
plt.grid(True)
plt.legend()
plt.show()
```

# 5)Sigmoidal MF

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Sigmoidal membership function
def sigmoid_mf(x, a, c):
    return 1 / (1 + np.exp(-a * (x - c)))

# Define the x-axis range
x = np.linspace(0, 12, 1000)

# Parameters
a = 2     # Slope
c = 6     # Center

# Compute membership values
y = sigmoid_mf(x, a, c)

# Plot
plt.plot(x, y, label=f'Sigmoidal MF (a={a}, c={c})', color='red')
plt.title('Sigmoidal Membership Function')
plt.xlabel('Input')
plt.ylabel('Membership Degree')
plt.grid(True)
plt.legend()
plt.show()
```

# 6) S shaped MF

```python
import numpy as np
import matplotlib.pyplot as plt

# Define S-function membership
def s_function(x, a, b, c):
    y = np.zeros_like(x)
    for i in range(len(x)):
        if x[i] <= a:
            y[i] = 0
        elif a < x[i] <= b:
            y[i] = 2 * ((x[i] - a) / (c - a)) ** 2
        elif b < x[i] < c:
            y[i] = 1 - 2 * ((c - x[i]) / (c - a)) ** 2
        else:
            y[i] = 1
    return y

# Define x-axis
x = np.linspace(0, 12, 1000)

# Parameters
a, b, c = 2, 6, 10

# Compute S-function
y = s_function(x, a, b, c)

# Plot
plt.plot(x, y, label=f'S-function (a={a}, b={b}, c={c})', color='blue')
plt.title('S-function Membership Function')
plt.xlabel('Input')
plt.ylabel('Membership Degree')
plt.grid(True)
plt.legend()
plt.show()
```

# 7) constant function

```python
import numpy as np
import matplotlib.pyplot as plt

# Define constant membership function
def constant_mf(x, k):
    return np.full_like(x, k)

# x-axis range
x = np.linspace(0, 12, 1000)

# Constant value
k = 0.7  # You can set this to 0, 0.5, or 1 too

# Compute membership values
y = constant_mf(x, k)

# Plot
plt.plot(x, y, label=f'Constant MF (k={k})', color='brown')
plt.title('Constant Membership Function')
plt.xlabel('Input')
plt.ylabel('Membership Degree')
plt.ylim(-0.1, 1.1)
plt.grid(True)
plt.legend()
plt.show()
```
